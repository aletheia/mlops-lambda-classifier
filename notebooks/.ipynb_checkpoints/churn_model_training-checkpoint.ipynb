# Classify churners from credit card account records

For this example we consider using an existing dataset available on Kaggle at url (https://www.kaggle.com/sakshigoyal7/credit-card-customers) which presents a set of records related to bank accounts. Our goal is to predict the churners of a single record

```python
# include numpy and pandas to handle linear algebra and dataset manipulation
import numpy as np 
import pandas as pd 
```

```python
# Dataset has been downloaded and placed in the same folder of this notebook with name BankChurners.csv.

dataset_name = './BankChurners.csv'
```

```python
df_raw = pd.read_csv(dataset_name)
df_raw.head()
```

## Feature analysis

```python
# Let's print the name of the colums to have an idea of the data
df_raw.columns.values
```

```python
# Let's print field values

print(df_raw['Attrition_Flag'].value_counts())
print(df_raw['Gender'].value_counts())
print(df_raw['Education_Level'].value_counts())
print(df_raw['Marital_Status'].value_counts())
print(df_raw['Income_Category'].value_counts())
print(df_raw['Card_Category'].value_counts())
```

```python
# Feature encoding
mapping = {
    'Attrition_Flag': {
        'Existing Customer' : 0, 
        'Attrited Customer' : 1
    },
    'Gender': {
        'F': 0, 
        'M': 1
    },
    'Education_Level': {
        'Unknown' : 0, 
        'Uneducated' : 1, 
        'High School' : 2, 
        'College' : 3, 
        'Graduate' : 4, 
        'Post-Graduate' : 5, 
        'Doctorate' : 6
    },
    'Marital_Status': {
        'Unknown' : 0, 
        'Single' : 1, 
        'Divorced' : 2, 
        'Married' : 3
    },
    'Income_Category': {
        'Unknown' : 0, 
        'Less than $40K' : 1, 
        '$40K - $60K' : 2, 
        '$60K - $80K' : 3,
        '$80K - $120K' : 4, 
        '$120K +' : 5
    },
    'Card_Category': {
        'Blue' : 0, 
        'Silver' : 1, 
        'Gold' : 2, 
        'Platinum' : 3
    }
}
```

```python
# Loop through mappings and apply feature encoding 
df = df_raw.copy()
for f in mapping:
    df[f].replace(mapping[f], inplace=True)
    
# Drop unuseful columns
df = df.drop(['Unnamed: 0','CLIENTNUM'], axis=1)

# Print new dataset
df.head()
```

```python
# Print types
df.dtypes
```

### Dimensional reduction

Now we remove feature that have a very weak correlation with attrition because they represent just underlying noise in our dataset. Feature correlation analysis is a fundamental step that can be done manually or automatically using PCA/TSNE. After related feature have been discovered (if any) we can remove uncorrelated features, to enhance classification quality.

```python
# Print correlations
df.corr()
```

The most correlated fields to churn are transaction counts, count change from Q4 to Q1, revolving balance, 12 month contact count, inactive months, utilization ratio,relationship count, and transaction amount. That being said, none of them share a particularly strong correlation, but essentially all of the demographic information (gender, income, education level, marital status, dependant count) lack any real correlation with churn.

```python
# Define a new dataset keeping only features that have a decent correlation with Attrition Flag, 
# which is representing churn
dfm = df[['Attrition_Flag', 'Total_Relationship_Count', 'Months_Inactive_12_mon',
          'Contacts_Count_12_mon', 'Total_Revolving_Bal', 'Total_Trans_Amt', 'Total_Trans_Ct',
          'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']]
dfm.corr()
```

## Model creation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
x = dfm[['Total_Relationship_Count', 'Months_Inactive_12_mon', 'Contacts_Count_12_mon',
        'Total_Revolving_Bal', 'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
        'Avg_Utilization_Ratio']]
y = dfm['Attrition_Flag']

# Create training dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
# Model training
model = RandomForestClassifier(n_estimators=100, max_depth=13, random_state=2)
model.fit(x, y)
rfvalue = model.predict(x_test)

print('Model Accuracy : ', accuracy_score(y_test, rfvalue) *  100)
print('Model Recall : ', recall_score(y_test, rfvalue) *  100)
print('Model Precision : ', precision_score(y_test, rfvalue) *  100)
```

```python
# Save the model
import pickle
filename = 'trained_model.mdl'
pickle.dump(model, open(filename, 'wb'))
```

## Analyze model results

```python
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
```

```python
df_raw['Attrition_Flag'].unique()
```

```python
class_names = df_raw['Attrition_Flag'].unique()
normalize = 'true'
title = 'Confusion matrix with normalization'
disp = plot_confusion_matrix(
    model, x_test, y_test, display_labels=class_names, cmap=plt.cm.Blues, normalize=normalize)
disp.ax_.set_title(title)
#print(title)
#print(disp.confusion_matrix)
```

```python
normalize = None
title = "Confusion matrix without normalization"
disp = plot_confusion_matrix(
    model, x_test, y_test, display_labels=class_names, cmap=plt.cm.Blues, normalize=normalize)
disp.ax_.set_title(title)
```
